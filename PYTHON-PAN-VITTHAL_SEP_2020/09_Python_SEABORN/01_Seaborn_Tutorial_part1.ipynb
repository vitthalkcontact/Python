##### Swarm plot

```python
sns.violinplot(df.acceleration,orient="v",color="g")
sns.swarmplot(df.acceleration,orient="v",color="r")
plt.show()
```

```python
sns.violinplot(x=df.origin,y=df.acceleration,orient="v",color="g")
sns.swarmplot(x=df.origin,y=df.acceleration,orient="v",color="r",hue=df.cylinders,alpha=.2)
plt.show()
```

```python
sns.violinplot(x=df.origin,y=df.acceleration,orient="v",color="g",inner="quartile")
sns.swarmplot(x=df.origin,y=df.acceleration,orient="v",color="r",hue=df.cylinders,alpha=.2)
plt.show()
```

##### category(cat) plot

```python
df = sns.load_dataset("tips")
df.head(2)
```

```python
sns.catplot(x="size",y="total_bill",row="smoker",col="time",data=df,hue="day")
```

```python
sns.catplot(x="size",y="total_bill",row="smoker",col="time",data=df,height=3,aspect=2,hue="day")
```

```python
sns.catplot(x="size",y="total_bill",row="smoker",col="day",data=df,height=3,aspect=1,hue="time")
```

```python
sns.catplot(x="size",y="total_bill",col_wrap=2,col="day",data=df,height=3,aspect=1,hue="time")
```

##### HeatmapPlot

```python
dia = sns.load_dataset("diamonds")
```

```python
dia.head(2)
```

```python
dia.shape
```

```python
cor = dia.corr()
```

```python
cor
```

```python
sns.heatmap(cor)
```

```python
sns.heatmap(cor,annot=True)
```

```python
sns.heatmap(cor,annot=True,fmt=".2f")
```

```python
sns.heatmap(cor,annot=True,fmt=".2f",cmap="jet_r")
```

```python
sns.heatmap(cor,annot=True,fmt=".2f",cmap="jet_r",cbar=False)
```

```python
sns.heatmap(cor,annot=True,fmt=".2f",cmap="jet_r",cbar=False,linewidths=.2)
```

```python
sns.heatmap(cor,annot=True,fmt=".2f",cmap="jet_r",cbar=False,linewidths=.2,linecolor="r")
```

```python
sns.heatmap(cor,annot=True,fmt=".2f",cmap="jet_r",vmin=.5,vmax=.75)
```

```python
sns.heatmap(cor,annot=True,fmt=".2f",cmap="jet_r",vmin=.95,vmax=1.0)
```

```python
sns.heatmap(cor,annot=True,fmt=".2f",cmap="winter")
```

```python
sns.heatmap(cor,annot=True,fmt=".2f",cmap="winter",vmin=.95,vmax=1.0)
```
