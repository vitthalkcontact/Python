# Dictionaries
```
Dictionaries store connections between pieces of information. Each item in a dictionary is a key-value pair.

Dictinaries are unordered mappings for storing objects
```

````python
# A simple dictionary
alien = {'color': 'green', 'points': 5}
````

````python
# Accessing a value
print("The alien's color is " + alien['color'])
````

````python
# Adding a new key-value pair
alien['x_position'] = 0
alien
````

````python
# Looping through all key-value pairs
fav_numbers = {'eric': 17, 'ever': 4} 
for name, number in fav_numbers.items():
    print(name + ' loves ' + str(number))
````

````python
som = {'ali':4, 'bal':10}
for nm, nr in som.items():
    print(nm + ' likes'+ ' '+ str(nr))
````

````python
# Looping through all keys
fav_numbers = {'eric': 17, 'ever': 4} 
for name in fav_numbers.keys():
    print(name + ' loves a number')
````

````python
# Looping through all the values
fav_numbers = {'eric': 17, 'ever': 4} 
for number in fav_numbers.values():
    print(str(number) + ' is a favorite')
````

# Error Handling and Eceptions

# Syntaxt
```
try:
    #You attempt your operation here
except Exception-I:
    #If Exception-I, then execute this block
except Exception-II:
    #If Exception-II, then execute this block
else:
    #If there is no Exception, then execute this block
finally:
    #Excute this block
```

````python
try:
    result = input ("Enter a number")
    1+int(result)
except:
    print('Hey enter a correct number')
else:
    print('Everting is alright')
finally:
    print('OK')
````
