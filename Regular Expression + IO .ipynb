```python
### regex

import re

mystring = 'Cats are better than dogs'
#### 

### match (looks into only first word of the string), 
### search (searches anywhere within the given string)

t = re.match(pattern = 'cats', string = mystring, flags = re.I)
t.group()
```

```python
s = re.search(pattern = 'than dog', string = mystring, flags = re.I)
s.group()
```

```python
s.groups(1)
```

```python
m = re.search('(?<=abc)def', 'abcdef def1 abc def')
m
```

```python
m.group()
```

```python
str = 'an example word:cat!!'
match = re.search(r'\w\w\w', str)
# If-statement after search() tests if it succeeded
if match:
    print ('found', match.group()) ## 'found word:cat'
else:
    print ('did not find')
```

```python
###
cp = re.compile(pattern = 'cats')
```

```python
cp.search('cats are preety good.').group()
```

```python
cp.search('cats are bad.').group()
```

# input output 

 - open(), close()
 - write()
 - writelines()

```python
file = open('file.txt','a')
# file.write('This is first line. \nThis is second line.')
file.write('\nThis is third line. \nThis is fourth line.')
# file.writelines()
file.close()
```

```python
file = open('file.txt','a')
# file.write('This is first line. \nThis is second line.')
file.writelines(['\nThis is third line. \nThis is fourth line.'])
# file.writelines()
file.close()
```

## read methods
 - read()
 - readline()
 - readlines()

```python
file = open('file.txt','r')
print(file.read())
file.close()
```

```python
file = open('file.txt','r')
print(file.readline())
print(file.readline())
file.close()
```

```python
file = open('file.txt','r')
print(file.readlines())
file.close()
```

```python
import urllib.request
infile = urllib.request.urlopen('http://www.yahoo.com')
print(infile.read())
```

```python
import urllib.request
infile = urllib.request.urlopen('http://www.yahoo.com')
print(infile.read().decode())
```
