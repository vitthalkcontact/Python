# Matplotlib

# plotting library for the Python

```python
# importing matplotlib
from matplotlib import pyplot as plt 
```

```python
# x-axis values 
x = [15, 12, 19, 34, 67] 
  
# Y-axis values 
y = [10, 15, 18, 14, 12] 
  
# Function to plot 
plt.plot(x,y) 
  
# function to show the plot 
plt.show() 
```

```python
# Function to plot the bar 
plt.bar(x,y) 
```

```python
# Function to plot histogram 
plt.hist(y) 
```

```python
# Function to plot scatter 
plt.scatter(x, y)
```
