# Prime Number

````python
def is_prime(number):
    for each in range (2,number):
        if number % each == 0:
            return False
    return True  
````

````python
def is_prime(number):
    for each in range(2,number): # run a loop from 2 to number - 1
        if number % each == 0: # check if its getting divide
            return False # if `yes, its not prime
    
    return True # not divided by any number, its a prime
````

````python
# find all prime between 10 and 15
for each in range(10,15):
    print(each, is_prime(each))
````

# File Handling

- Below I have illustrated 3 ways to open and read a file.
- Please go thru each example and understand them carefully.
- Make sure you have log.txt file in the same folder where you are executing the code


### Caution

- If you get ``` FileNotFoundError: [Errno 2] No such file or directory: 'log.txt' ``` ERROR.
- This means it can not find log.txt in current folder

# 1. Read all data at once, and print

````python
file = open("log.txt")
# all data at once
data = file.read()
print(data)
file.close()
````

# 2. Read all data at once, and print using with open

````python
with open("log.txt") as file:
    # all data at once
    data = file.read()
    print(data)
````

# 3. Read line by line, and print

````python
file = open('log.txt')
for each_line in file:
    print('****************** A new line ******************')
    print(each_line)
    
file.close()
````

# 4. Read all data at once, and print using with open

````python
with open("log.txt") as file:
    # line by line
    for each_line in file:
        print('****************** A new line ******************')
        print(each_line)
````
